```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the spotify Disease dataset
df = pd.read_csv('top_50_2023.csv', encoding='latin1')

# Display the first few rows of the dataset
print(df.head())

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Handle non-numeric values
df_numeric = df.apply(pd.to_numeric, errors='coerce')
#df_numeric.dropna(inplace=True)

# Display basic statistics of the dataset
# print(df_numeric.describe())

# Calculate and display major moments for numerical features
moments = df_numeric.describe().transpose()[['mean', '50%', 'std']]
skewness = df_numeric.skew()
kurtosis = df_numeric.kurtosis()

# Combine moments, skewness, and kurtosis into a single DataFrame
statistics_df = pd.concat([moments, skewness, kurtosis], axis=1)
statistics_df.columns = ['Mean', 'Median', 'Standard Deviation', 'Skewness', 'Kurtosis']

# Display the statistics
print(statistics_df)
```

```python
# Compute correlation matrix
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
```

```python
# Create histogram
plt.hist(df_numeric['popularity'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Popularity')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create scatter plot
plt.scatter(df_numeric['danceability'], df_numeric['popularity'], color='orange')
plt.title('Relationship between Danceability and Popularity')
plt.xlabel('Dance')
plt.ylabel('Popular')
plt.show()
```

```python
# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
