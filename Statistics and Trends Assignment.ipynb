```python
import pandas as pd

# Assuming the file is in the specified directory
file_path = "~/Desktop/Statistics and Trends Assignment/top_50_2023.csv"

# Importing the dataset into a DataFrame
df = pd.read_csv(file_path)

# Displaying the first few rows of the DataFrame to verify the import
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Assuming 'artist_name' and 'popularity' are the relevant columns in your DataFrame
top_artists = df.groupby('artist_name')['popularity'].max().nlargest(10)

# Plotting the bar graph
plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Popular Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sort the DataFrame by 'popularity' column in descending order and select the top 15 rows
top_15_artists = df.sort_values(by='popularity', ascending=False).head(15)

# Plotting the bar graph
plt.figure(figsize=(10, 8))
plt.barh(top_15_artists['artist_name'], top_15_artists['popularity'], color='skyblue')
plt.xlabel('Popularity')
plt.ylabel('Artist')
plt.title('Top 15 Artists Based on Popularity')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest popularity at the top
plt.show()

```

```python
import matplotlib.pyplot as plt

# Assuming 'duration_ms' is the column containing song durations in milliseconds
plt.figure(figsize=(10, 6))
plt.hist(df['duration_ms'] / 60000, bins=20, color='skyblue', edgecolor='black')  # Convert milliseconds to minutes
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Song Durations')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Assuming you have loaded your dataset into a pandas DataFrame named df
# and the duration of each song is in the column 'duration_ms'

# Calculate the average duration
average_duration_ms = df['duration_ms'].mean()

# Convert milliseconds to minutes and seconds
average_duration_min = average_duration_ms / 60000
average_duration_sec = (average_duration_min % 1) * 60

print("Average duration of songs: {:.0f} minutes and {:.0f} seconds".format(average_duration_min, average_duration_sec))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and standard deviation
mean_duration = 200  # 3 minutes and 20 seconds in seconds
std_dev_duration = 20  # Adjust this value to control the spread of the distribution

# Generate random samples following a normal distribution
samples = np.random.normal(mean_duration, std_dev_duration, 1000)

# Plot the histogram
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Add a vertical line at the mean
plt.axvline(mean_duration, color='r', linestyle='dashed', linewidth=1)

# Add labels and title
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.title('Normal Distribution of Song Durations')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named 'df' and the 'genres' column contains the list of genres for each song
# First, you may need to preprocess the 'genres' column to separate individual genres

# Convert the string representation of list to actual list
df['genres'] = df['genres'].apply(eval)

# Create a list to store all genres and corresponding artists
genre_artist_dict = {}

# Iterate through each row and extract genres and artists
for index, row in df.iterrows():
    for genre in row['genres']:
        if genre in genre_artist_dict:
            genre_artist_dict[genre].append(row['artist_name'])
        else:
            genre_artist_dict[genre] = [row['artist_name']]

# Count the number of unique artists for each genre
genre_artist_count = {genre: len(set(artists)) for genre, artists in genre_artist_dict.items()}

# Convert the dictionary to a DataFrame for plotting
genre_artist_df = pd.DataFrame(list(genre_artist_count.items()), columns=['Genre', 'Number of Artists'])

# Sort the DataFrame by the number of artists
genre_artist_df = genre_artist_df.sort_values(by='Number of Artists', ascending=False)

# Plot the bar chart
plt.figure(figsize=(12, 8))
plt.bar(genre_artist_df['Genre'], genre_artist_df['Number of Artists'], color='skyblue')
plt.title('Number of Artists by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Artists')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named 'df' and the 'genres' column contains the list of genres for each song
# First, you may need to preprocess the 'genres' column to separate individual genres

# Convert the string representation of list to actual list
df['genres'] = df['genres'].apply(eval)

# Create a list to store all genres and corresponding artists
genre_artist_dict = {}

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Extract the genres list for the current row
    genres_list = row['genres']
    
    # Iterate over each genre in the genres list
    for genre in genres_list:
        # Check if the genre exists in the dictionary
        if genre in genre_artist_dict:
            # Append the artist name to the list of artists for the current genre
            genre_artist_dict[genre].append(row['artist_name'])
        else:
            # Create a new key-value pair in the dictionary with the genre as the key and a list containing the artist name as the value
            genre_artist_dict[genre] = [row['artist_name']]

# Create lists to store genres and corresponding number of artists
genres = []
num_artists = []

# Iterate over the dictionary items and append genre and number of artists to the lists
for genre, artists in genre_artist_dict.items():
    genres.append(genre)
    num_artists.append(len(artists))

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.barh(genres, num_artists, color='skyblue')
plt.xlabel('Number of Artists')
plt.ylabel('Genres')
plt.title('Number of Artists by Genre')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary to store genres and corresponding artists
genre_artist_dict = {}

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Extract the genres list for the current row
    genres_list = row['genres']
    
    # Iterate over each genre in the genres list
    for genre in genres_list:
        # Check if the genre exists in the dictionary
        if genre in genre_artist_dict:
            # Append the artist name to the list of artists for the current genre
            genre_artist_dict[genre].append(row['artist_name'])
        else:
            # Create a new key-value pair in the dictionary with the genre as the key and a list containing the artist name as the value
            genre_artist_dict[genre] = [row['artist_name']]

# Create lists to store genres and corresponding list of artists
genres = []
artists_list = []

# Iterate over the dictionary items and append genre and list of artists to the lists
for genre, artists in genre_artist_dict.items():
    genres.append(genre)
    artists_list.append(', '.join(artists))  # Join the list of artists into a single string separated by commas

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(genres, artists_list, color='skyblue')
plt.xlabel('Genres')
plt.ylabel('Artists')
plt.title('Artists by Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
